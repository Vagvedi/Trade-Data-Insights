```python
df_trades = pd.read_csv("historical_data.csv")
def parse_timestamp(ts):
    try:
        ts_int = int(ts)
        if len(str(ts_int)) == 13:
            return pd.to_datetime(ts_int, unit='ms')
        elif len(str(ts_int)) == 10:
            return pd.to_datetime(ts_int, unit='s')
        else:
            return pd.NaT
    except:
        return pd.NaT

df_trades['Timestamp'] = df_trades['Timestamp'].apply(parse_timestamp)
df_trades['date'] = df_trades['Timestamp'].dt.date
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df_trades = pd.read_csv("historical_data.csv")
df_sentiment = pd.read_csv("fear_greed_index.csv")

# Preprocessing
df_trades['Timestamp'] = pd.to_datetime(df_trades['Timestamp'], errors='coerce')
df_trades['Closed PnL'] = pd.to_numeric(df_trades['Closed PnL'], errors='coerce')
df_trades['Size USD'] = pd.to_numeric(df_trades['Size USD'], errors='coerce')

# Filter out bad rows BEFORE merging
df_trades = df_trades.dropna(subset=['Timestamp', 'Closed PnL', 'Size USD'])
df_trades = df_trades[(df_trades['Closed PnL'] != 0) & (df_trades['Size USD'] != 0)]

# Format date
df_trades['date'] = df_trades['Timestamp'].dt.date
df_sentiment['date'] = pd.to_datetime(df_sentiment['date'], errors='coerce').dt.date

# Merge
df_merged = pd.merge(df_trades, df_sentiment[['date', 'classification']], on='date', how='left')

# Check sentiment distribution
print(df_merged['classification'].value_counts(dropna=False))

# Final clean filtered dataset
df_filtered = df_merged.dropna(subset=['classification', 'Size USD', 'Closed PnL'])

# Check that values exist
print(df_filtered[['Size USD', 'Closed PnL']].describe())


```

```python
 #  Scatter Plot: Trade Size vs Profit/Loss by Sentiment

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df_filtered,
    x='Size USD',
    y='Closed PnL',
    hue='classification',
    alpha=0.6,
    palette='coolwarm'
)

plt.title('Trade Size vs Profit/Loss by Market Sentiment', fontsize=14, weight='bold')
plt.xlabel('Trade Size (USD)', fontsize=12)
plt.ylabel('Closed Profit/Loss', fontsize=12)
plt.grid(True)
plt.legend(title='Market Sentiment')
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap of Numerical Features

import seaborn as sns
import matplotlib.pyplot as plt


numeric_cols = ['Execution Price', 'Size USD', 'Closed PnL', 'Fee']
df_numeric = df_filtered[numeric_cols].apply(pd.to_numeric, errors='coerce')

plt.figure(figsize=(10, 6))
corr_matrix = df_numeric.corr()

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="YlGnBu",
    linewidths=0.5,
    square=True
)

plt.title(' Correlation Heatmap of Trade Metrics', fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

```python
# Pairplot: Trade Size vs PnL vs Fee (Colored by Sentiment)

import seaborn as sns
import matplotlib.pyplot as plt

df_sample = df_filtered[['Size USD', 'Closed PnL', 'Fee', 'classification']].dropna().sample(n=1000, random_state=1)

sns.pairplot(
    df_sample,
    hue='classification',
    palette='husl',
    diag_kind='kde',
    plot_kws={'alpha': 0.6, 's': 40}
)

plt.suptitle('Pairplot of Trade Metrics by Market Sentiment', fontsize=16, weight='bold', y=1.02)
plt.show()
```

```python
# Correlation Heatmap of Trade Metrics

import seaborn as sns
import matplotlib.pyplot as plt
numeric_cols = ['Execution Price', 'Size Tokens', 'Size USD', 'Fee', 'Closed PnL', 'Start Position']
df_corr = df_filtered[numeric_cols].apply(pd.to_numeric, errors='coerce').dropna()

corr_matrix = df_corr.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, cbar_kws={'shrink': .8})
plt.title(' Correlation Heatmap of Trade Metrics', fontsize=16, weight='bold')
plt.tight_layout()
plt.show()
```

```python
# Boxen Plot: Closed PnL Distribution by Sentiment

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.boxenplot(
    data=df_filtered,
    x='classification',
    y='Closed PnL'
)

plt.title('Distribution of Profit/Loss by Market Sentiment', fontsize=16, weight='bold')
plt.xlabel('Market Sentiment')
plt.ylabel('Closed Profit/Loss')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Time Series Plot: Avg Daily PnL by Market Sentiment

import matplotlib.pyplot as plt
import seaborn as sns

df_filtered['date'] = pd.to_datetime(df_filtered['date'], errors='coerce')

daily_pnl = df_filtered.groupby(['date', 'classification'])['Closed PnL'].mean().reset_index()

plt.figure(figsize=(14, 6))
sns.lineplot(
    data=daily_pnl,
    x='date',
    y='Closed PnL',
    hue='classification',
    palette='Set1'
)

plt.title('Avg Daily Profit/Loss by Market Sentiment', fontsize=16, weight='bold')
plt.xlabel('Date')
plt.ylabel('Avg Closed PnL')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output

df_filtered['date'] = pd.to_datetime(df_filtered['date'], errors='coerce')
df_filtered_sorted = df_filtered.sort_values('date')

pnl_daily = df_filtered_sorted.groupby('date')['Closed PnL'].mean().reset_index()


for i in range(20, len(pnl_daily), 10):
    clear_output(wait=True)
    plt.figure(figsize=(12, 5))
    sns.lineplot(x='date', y='Closed PnL', data=pnl_daily[:i], color='green')
    plt.title(' Simulated Live Plot: Avg Daily Closed PnL', fontsize=16, weight='bold')
    plt.xlabel('Date')
    plt.ylabel('Avg Closed PnL')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    time.sleep(0.5)
```

```python
# Violin + Stripplot Combo Plot: Closed PnL Spread by Sentiment

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))


sns.violinplot(
    data=df_filtered,
    x='classification',
    y='Closed PnL',
    inner=None,
    color='skyblue',
    cut=0
)

sns.stripplot(
    data=df_filtered.sample(400, random_state=42),
    x='classification',
    y='Closed PnL',
    color='black',
    alpha=0.4,
    size=2.5
)

plt.title(' Final Combo: PnL Spread with Data Points by Market Sentiment', fontsize=16, weight='bold')
plt.xlabel('Market Sentiment')
plt.ylabel('Closed Profit/Loss')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

```
